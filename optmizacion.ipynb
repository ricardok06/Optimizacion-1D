##Polynomial neither Odd nor Even functions


```python
import matplotlib.pyplot as plt
import numpy as np
a = -5
b = 5#intervalo
x = np.linspace(a, b, 100)
fx = '-x**6/60-x**5/50+x**4/2+2*x**3/3-3.2*x**2-6.4*x'
plt.figure()
plt.plot(x, eval(fx))#graficamos la figura, usamos eval, para usar la funcion fx
plt.grid()
plt.show()
```

###Programa

```python
import sympy #importamos sympy para hacer la derivada
def f(x):
    return eval(fx) #evaluamos la funcion en x
def d1f(x):
    return eval(str(sympy.diff(fx))) #hacemos la derivada de la funcion, convertimos a cadena y evaluamos
```

```python
def biseccion(a,b,R):
    epsilon = 0.001 #paro
    cont = 0
    registro = []

    while(True):# el while continuara hasta que se cumpla la condicion en el if
        alpha = (a+b)/2

        # Primera derivada en a
        df_a = d1f(a)

        #Primera derivada en alpha
        df_alpha = d1f(alpha)

        if(df_a*df_alpha < 0):
            b = alpha
        else:
            a = alpha

        fa = f(a)

        # Condicion de finalizacion
        if(np.abs(a-b) < epsilon):
            print('-------------------------------------------------------------')
            print("It: {:02} - x: {:.10f} - f(x) {:.10f}".format(cont, a, fa))
            R.append(a)
            R.append(fa)
            break

        cont = cont + 1
        registro.append([cont, a, fa])
        print("It: {:02} - x: {:.10f} - f(x) {:.10f}".format(cont, a, fa))

    return registro

R=[]
reg1 = biseccion(a,b,R)
print(R)
```

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(R[0], color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('Temperatura')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(R[1], color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Costo')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()


evaluacion(reg1)
```

El metodo de la biseccion ha a lo mas en 8 iteraciones el maximo

```python
import sympy as sp
x = sp.symbols('x')
expresion = -x**4 * (sp.log(sp.Abs(x)) - 1)
derivada=sp.diff(expresion)
r= sp.simplify(derivada) #simplificamos la expresion
print(r)
```

```python
def d1f(x_real):
    # Definir el smbolo x
    x = sp.symbols('x')

    # Definir las partes real e imaginaria de x
    re_x = sp.re(x)
    im_x = sp.im(x)

    # Definir la expresin
    expresion = x**3 * (-x * (re_x * sp.diff(re_x, x) + im_x * sp.diff(im_x, x)) + (4 - 4 * sp.log(sp.Abs(x))) * sp.Abs(x)**2) / sp.Abs(x)**2

    # Evaluar la expresin en el valor dado de x_real
    resultado = expresion.subs(x, x_real)

    return resultado
```

```python
def f(x):
  return -x**6/60-x**5/50+x**4/2+2*x**3/3-3.2*x**2-6.4*x
```

```python
def GoldenSectionSearch(a,b,R):
    tau = 2 - 1.618033988 #inicializamos tau
    epsilon = 1e-6

    cont = 0
    registro = []

    while(True):
        # Calcular alpha1 y alpha2
        alpha1 = a*(1 - tau) + b*tau
        alpha2 = a*tau + b*(1 - tau)

        # Calcular f(alpha1) y f(alpha2)
        f_alpha1 = f(alpha1)
        f_alpha2 = f(alpha2)

        if(f_alpha1 > f_alpha2):
            a = alpha1
        else:
            b = alpha2

        cont = cont + 1
        registro.append([cont, alpha1, f_alpha1])
        print("It: {:02d} - x: {:.10f} - f(x): {:.10f}".format(cont, alpha1, f_alpha1))

        if(np.abs(f_alpha1 - f_alpha2) < epsilon):
            print("-------------------------------------------------------")
            print("It: {:02d} - x: {:.10f} - f(x): {:.10f}".format(cont, alpha1, f_alpha1))
            R.append(alpha1)
            R.append(f_alpha1)
            break

    return registro
R=[]
reg = GoldenSectionSearch(a,b,R)
print(R)
```

```python
def evaluacion(reg):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(1.999, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('xemperatura')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(-13.97, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Costo')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()
evaluacion(reg)
```

```python
def Fibonacci(*params):#pasamos elementos arbitrarios
    a   = params[0]#numero del parametro agregado
    b   = params[1]
    k   = params[2]
    it  = params[3]
    reg = params[4]
    L   = params[5]
    n   = params[6]
    fs  = params[7]
    ep  = params[8]#variable de paro

    i_arriba = n - k + 1 #numerador
    i_abajo = n + 1   #denominador
    Lk = L*fs[i_arriba]/fs[i_abajo]

    xa = a + Lk
    xb = b - Lk

    f_xa = f(xa)
    f_xb = f(xb)

    if(f_xa > f_xb):
        a = xa
    else:
        b = xb

    if(np.abs(f_xa-f_xb) > ep):
        k = k + 1
        it = it + 1

        reg.append([it, xa, f_xa])
        print("It: {:02d} - xemp: {:.10f} - Costo: {:.10f}".format(it, xa, f_xa))

        return Fibonacci(a, b, k, it, reg, L, n, fs, ep)
    else:
        print("-------------------------------------------------------")
        print("It: {:02d} - xemp: {:.10f} - Costo: {:.10f}".format(it, xa, f_xa))

        return reg
```

```python
def RunFibonacci(a,b):#serie de fibonacci
    n = 30 #ecantidad elementos de la serie

    fs = [1, 1] #elementos de la serie de fibonacci
    for i in range(n):
        aux = fs[-1] + fs[-2] #serie
        fs.append(aux)

    k =2
    L = b - a #definimos L

    return Fibonacci(a, b, k, 0, [], b-a, n, fs, 1e-6) #registro
reg = RunFibonacci(a,b)
```

```python
def evaluacion(reg,o,t):
    reg = np.array(reg)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    fig.suptitle('Analisis de convergencia')

    axs[0].axhline(o, color='k', linewidth=3, linestyle='--')
    axs[0].plot(reg[:, 0], reg[:, 1], linewidth=4)
    axs[0].set_xlabel('Iteraciones')
    axs[0].set_ylabel('xemperatura')
    axs[0].grid()
    axs[0].set_xlim([0, 15])

    axs[1].axhline(t, color='k', linewidth=3, linestyle='--')
    axs[1].plot(reg[:, 0], reg[:, 2], linewidth=4)
    axs[1].set_xlabel('Iteraciones')
    axs[1].set_ylabel('Costo')
    axs[1].grid()
    axs[1].set_xlim([0, 15])

    plt.show()
evaluacion(reg,1.99,-13.97)
```
